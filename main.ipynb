```python
# Specify the directory and the value of n
paths = ["f22", "w23", "s23", "f23", "w24"]
n_value = 8
homePath = os.getcwd()
classes = defaultdict(list) # keep track of classes
```

```python
def extract_last_n_chars(directory, n):
    %cd $directory

    # Loop through all files in the directory
    for filename in os.listdir('.'):
        # Check if the item is a file (not a directory)
        if os.path.isfile(filename):
            # Extract up to last n characters (class name/num)
            extracted = filename[:-n]
            quarter = filename[-7:-4]

            num = "" # keep track of course number
            for i in range(3, len(extracted)):
                if extracted[i].isnumeric():
                    num += extracted[i]
                else:
                    break

            
            if num != '':
                course = int(num)

                if course > 200: # found a graduate course
                    os.remove(filename) # remove all grad classes
                    continue

            # valid class; process as df
            if extracted != '':
                curr = pd.read_csv(filename)
                
                classes[extracted].append((quarter, curr))

    %cd ../
```

```python
# Call the function
for p in paths:
    extract_last_n_chars(p, n_value)
```

```python
from datetime import datetime

# Convert the string representations to datetime objects
time_str1 = "2022-05-23T20:50:19"
time_str2 = "2022-05-24T10:30:45"

datetime1 = datetime.fromisoformat(time_str1)
datetime2 = datetime.fromisoformat(time_str2)

# Calculate the time difference
time_difference = datetime2 - datetime1

print(f"The time difference is: {time_difference}")
```

```python
hours, remainder = divmod(time_difference.seconds, 3600)
minutes, seconds = divmod(remainder, 60)

print(f"The time difference is: {time_difference.days} days, {hours} hours, {minutes} minutes, {seconds} seconds.")
```
